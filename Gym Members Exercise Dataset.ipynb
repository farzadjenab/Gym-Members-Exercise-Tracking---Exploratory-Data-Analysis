```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options for better visualization
pd.set_option('display.max_columns', None)

# Set style for plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# For displaying plots inline in Jupyter
%matplotlib inline

print("Libraries imported successfully! ✅")
```

```python
# Load the dataset
df = pd.read_csv('gym_members_exercise_tracking.csv')

# Display first 5 rows to see what we have
print("First 5 rows of the dataset:")
df.head()
```

```python
# Check the shape of dataset (rows, columns)
print(f"Dataset Shape: {df.shape}")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
# Check dataset shape
df.shape
```

```python
# Display column names
print("Column names in dataset:")
print(df.columns.tolist())
```

```python
# Check data types and non-null counts
print("Data types and info:")
df.info()
```

```python
# Get basic statistics for numerical columns
print("Statistical Summary of Numerical Columns:")
df.describe()
```

```python
# Round the statistics for better readability
df.describe().round(2)
```

```python
# Check for missing values in each column
print("Missing Values in Each Column:")
missing_values = df.isnull().sum()
print(missing_values)

# Calculate percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nMissing Values Percentage:")
print(missing_percentage)
```

```python
# Visualize missing values
plt.figure(figsize=(10, 5))
missing_values.plot(kind='bar', color='coral')
plt.title('Missing Values Count per Column', fontsize=14)
plt.xlabel('Columns')
plt.ylabel('Count of Missing Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Check unique values in categorical columns
print("Unique values in 'Gender':")
print(df['Gender'].value_counts())
print()
print("Unique values in 'Workout_Type':")
print(df['Workout_Type'].value_counts())
```

```python
# Visualize Gender distribution
plt.figure(figsize=(8, 5))
df['Gender'].value_counts().plot(kind='bar', color=['steelblue', 'salmon'])
plt.title('Gender Distribution', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Visualize Workout Type distribution
plt.figure(figsize=(10, 5))
df['Workout_Type'].value_counts().plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
plt.title('Workout Type Distribution', fontsize=14)
plt.xlabel('Workout Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Age distribution histogram
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=20, color='teal', edgecolor='white', alpha=0.7)
plt.title('Age Distribution of Gym Members', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(df['Age'].mean(), color='red', linestyle='--', label=f'Mean: {df["Age"].mean():.1f}')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Average Age: {df['Age'].mean():.1f} years")
print(f"Youngest Member: {df['Age'].min()} years")
print(f"Oldest Member: {df['Age'].max()} years")
```

```python
# Weight distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Weight (kg)'], bins=20, color='purple', edgecolor='white', alpha=0.7)
plt.title('Weight Distribution', fontsize=12)
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['BMI'], bins=20, color='orange', edgecolor='white', alpha=0.7)
plt.title('BMI Distribution', fontsize=12)
plt.xlabel('BMI')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# BMI categories
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# Apply the function
df['BMI_Category'] = df['BMI'].apply(categorize_bmi)

# Count BMI categories
print("BMI Category Distribution:")
print(df['BMI_Category'].value_counts())
```

```python
# Visualize BMI categories
plt.figure(figsize=(8, 5))
colors = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c']
df['BMI_Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=colors)
plt.title('BMI Category Distribution', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Calories burned analysis
plt.figure(figsize=(10, 5))
plt.hist(df['Calories_Burned'], bins=25, color='crimson', edgecolor='white', alpha=0.7)
plt.title('Calories Burned Distribution', fontsize=14)
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.axvline(df['Calories_Burned'].mean(), color='blue', linestyle='--', 
            label=f'Mean: {df["Calories_Burned"].mean():.0f}')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Average Calories Burned: {df['Calories_Burned'].mean():.0f}")
print(f"Min Calories Burned: {df['Calories_Burned'].min():.0f}")
print(f"Max Calories Burned: {df['Calories_Burned'].max():.0f}")
```

```python
# Calories burned by workout type
plt.figure(figsize=(10, 6))
df.groupby('Workout_Type')['Calories_Burned'].mean().sort_values(ascending=True).plot(
    kind='barh', color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
plt.title('Average Calories Burned by Workout Type', fontsize=14)
plt.xlabel('Average Calories Burned')
plt.ylabel('Workout Type')
plt.tight_layout()
plt.show()
```

```python
# Detailed statistics by workout type
print("Calories Burned Statistics by Workout Type:")
df.groupby('Workout_Type')['Calories_Burned'].agg(['mean', 'min', 'max', 'std']).round(2)
```

```python
# Compare average values by gender
gender_comparison = df.groupby('Gender')[['Calories_Burned', 'Session_Duration (hours)', 
                                           'Weight (kg)', 'BMI', 'Fat_Percentage']].mean().round(2)
print("Average Values by Gender:")
gender_comparison
```

```python
# Visualize gender comparison
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Calories Burned
df.groupby('Gender')['Calories_Burned'].mean().plot(kind='bar', ax=axes[0], color=['steelblue', 'salmon'])
axes[0].set_title('Avg Calories Burned by Gender')
axes[0].set_ylabel('Calories')
axes[0].set_xticklabels(['Female', 'Male'], rotation=0)

# Session Duration
df.groupby('Gender')['Session_Duration (hours)'].mean().plot(kind='bar', ax=axes[1], color=['steelblue', 'salmon'])
axes[1].set_title('Avg Session Duration by Gender')
axes[1].set_ylabel('Hours')
axes[1].set_xticklabels(['Female', 'Male'], rotation=0)

# Fat Percentage
df.groupby('Gender')['Fat_Percentage'].mean().plot(kind='bar', ax=axes[2], color=['steelblue', 'salmon'])
axes[2].set_title('Avg Fat Percentage by Gender')
axes[2].set_ylabel('Percentage')
axes[2].set_xticklabels(['Female', 'Male'], rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Session Duration vs Calories Burned
plt.figure(figsize=(10, 6))
plt.scatter(df['Session_Duration (hours)'], df['Calories_Burned'], 
            alpha=0.5, c='teal', edgecolors='white')
plt.title('Session Duration vs Calories Burned', fontsize=14)
plt.xlabel('Session Duration (hours)')
plt.ylabel('Calories Burned')
plt.tight_layout()
plt.show()

# Calculate correlation
correlation = df['Session_Duration (hours)'].corr(df['Calories_Burned'])
print(f"Correlation between Session Duration and Calories Burned: {correlation:.3f}")
```

```python
# Select numerical columns for correlation
numerical_cols = ['Age', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM', 
                  'Resting_BPM', 'Session_Duration (hours)', 'Calories_Burned', 
                  'Fat_Percentage', 'Water_Intake (liters)', 'Workout_Frequency (days/week)', 
                  'Experience_Level', 'BMI']

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', linewidths=0.5, square=True)
plt.title('Correlation Matrix of Numerical Variables', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Experience level distribution
plt.figure(figsize=(8, 5))
df['Experience_Level'].value_counts().sort_index().plot(kind='bar', 
                                                         color=['#3498db', '#2ecc71', '#e74c3c'])
plt.title('Experience Level Distribution', fontsize=14)
plt.xlabel('Experience Level (1=Beginner, 2=Intermediate, 3=Expert)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print("Experience Level Counts:")
print(df['Experience_Level'].value_counts().sort_index())
```

```python
# Compare metrics by experience level
experience_comparison = df.groupby('Experience_Level')[
    ['Calories_Burned', 'Session_Duration (hours)', 'Workout_Frequency (days/week)']
].mean().round(2)

print("Average Metrics by Experience Level:")
experience_comparison
```

```python
# Water intake analysis
plt.figure(figsize=(10, 5))
plt.hist(df['Water_Intake (liters)'], bins=15, color='deepskyblue', edgecolor='white', alpha=0.7)
plt.title('Water Intake Distribution', fontsize=14)
plt.xlabel('Water Intake (liters)')
plt.ylabel('Frequency')
plt.axvline(df['Water_Intake (liters)'].mean(), color='red', linestyle='--', 
            label=f'Mean: {df["Water_Intake (liters)"].mean():.2f} L')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Average Water Intake: {df['Water_Intake (liters)'].mean():.2f} liters")
```
